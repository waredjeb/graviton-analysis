```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['lines.color'] = 'r'
```

```python
path = "/home/wahid/Wahid/University/machine-learning/GG_HH_Gravi/"
file_graviton = path + "data_signal_gravi.csv"
file_ggh = path + "data_signal_GG_HH.csv"
file_ttb = path + "data_background_tt.csv"
```

```python
grav = pd.read_csv(file_graviton)
ggh = pd.read_csv(file_ggh)
ttb = pd.read_csv(file_ttb)
```

```python
grav.iloc[0,:]
```

```python
def plot_fast(label, bins):
    plt.figure(figsize = (10,8))
    grav.loc[:, label].plot.hist(bins = bins, label = 'grav', alpha = 0.3)
    ggh.loc[:, label].plot.hist(bins = bins, label = 'ggh', alpha = 0.3)
    ttb.loc[:, label].plot.hist(bins = bins, label = 'tt~', alpha = 0.3)
    plt.title(label)
    plt.legend()
```

```python
plot_fast("jet_max_pt", 100)
```

```python
plot_fast("jets_max_mass", 100)
```

```python
plot_fast("fat_jets_max_mass", 100)
```
